#### Handling Outlier

```python
#Outlier handler function

def outlier_handler(feature_name):
    
    Q1 = data[feature_name].quantile(0.25)
    Q3 =  data[feature_name].quantile(0.75)
    
    #Inter quantile range
    IQR = Q3-Q1
    
    #setting upper and lower limits 
    upper_limit = Q3+(1.5*IQR)
    lower_limit = Q1-(1.5*IQR)
    
    data[feature_name] = np.where(data[feature_name]>upper_limit,upper_limit,
                                 np.where(data[feature_name]<lower_limit,lower_limit,data[feature_name]))
```

```python
var = ['PM10', 'PM25', 'so2', 'no2']
for fname in var:
    outlier_handler(fname)
```

#### New feature creation

```python
def to_category(val):
    
    if val>0 and val<=50:
        return 'Good'
    
    if val>50 and val<=100:
        return 'Satisfactory'
    
    if val>100 and val<=200:
        return 'Moderate'
    
    if val>200 and val<=300:
        return 'Poor'
    
    if val>300 and val<=400:
        return 'Very Poor'
    
    if val>400:
        return 'Severe'
```

```python
data['AQI (Category Range)'] = data['AQI'].apply(to_category)
data.head()
```

```python
data['AQI (Category Range)'] = data['AQI (Category Range)'].map({'Good':0, 'Satisfactory':1,'Moderate':2,'Poor':3,'Very Poor':4,'Severe':5})
```

```python
data.nunique()
```

```python
# defining dependent and Independent variable
x = data.drop(['AQI', 'AQI (Category Range)'], axis = 1)
y = data['AQI']
```

```python
# splitting the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42, shuffle=True)
```

```python
#Applying Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

#### Regression

```python
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.neighbors import KNeighborsRegressor,KNeighborsClassifier
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
```

```python
#Linear Regression
linreg = LinearRegression()
linreg.fit(x_train_scaled, y_train)
```

```python
print('Accuracy on Training Data: ', linreg.score(x_train_scaled, y_train))
print("Accuracy on Testing Data: ", linreg.score(x_test_scaled, y_test))
```

```python
#Performance Metric
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

linregpred = linreg.predict(x_test_scaled)

print('Mean Absolute Error:', mean_absolute_error(y_test, linregpred))
print("Mean Squared Error: ", mean_squared_error(y_test, linregpred))
print("r2 score: ", r2_score(y_test, linregpred))

adjusted_r2_score = 1-(((1-r2_score(y_test, linregpred))*(x_test_scaled.shape[0]))/(x_test_scaled.shape[0]-x_test_scaled.shape[1]-1))

print('Adjusted r2 score: ', adjusted_r2_score)
```
